# Apunte de Python

### Tipado Dinámico
En Python no necesitas declarar el tipo de la variable al crearla, el intérprete lo asigna automáticamente según el valor.

Ejemplo:

```python
x = 10       # entero
x = "hola"   # ahora es cadena
```

La misma variable puede cambiar de tipo en tiempo de ejecución.

### Tipado Fuerte
Aunque sea dinámico, Python no convierte automáticamente los tipos de manera implícita si no tiene sentido.

Ejemplo:

```python
x = "10"
y = 5
print(x + y)   # Error: no se puede sumar str con int
```

En Python debes hacerlo explícito:

```python
print(int(x) + y)  # ✅ 15
```

### Entradas de datos con input()
Siempre devuelve una cadena (str), por lo que muchas veces necesitas convertir el tipo.

```python
nombre = input("Ingresa tu nombre: ")  # str
edad = int(input("Ingresa tu edad: ")) # convierte a int
altura = float(input("Ingresa tu altura: ")) # convierte a float

print(f"Hola {nombre}, tienes {edad} años y mides {altura} m.")
```

### Estructuras de control
Condicionales (if, elif, else)


```python
edad = int(input("Edad: "))

if edad >= 18:
    print("Eres mayor de edad")
elif edad > 12:
    print("Eres adolescente")
else:
    print("Eres niño")
```

### Bucles (for y while)


```python
# for recorre rangos o colecciones
for i in range(1, 6):
    print("Número:", i)

# while repite hasta que se cumpla una condición
contador = 0
while contador < 5:
    print("Contador:", contador)
    contador += 1
```

### Listas en Python
Colección ordenada, mutable y que permite elementos repetidos.

```python
frutas = ["manzana", "pera", "uva"]

frutas.append("plátano")   # agrega
frutas.remove("pera")      # elimina
print(frutas[0])           # acceso por índice
print(frutas[-1])          # último elemento
print(len(frutas))         # longitud
```
