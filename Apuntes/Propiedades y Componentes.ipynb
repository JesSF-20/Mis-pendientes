### 🧭 1. Variables globales (@)

 Estas definen la paleta base del tema;Todas pueden ser referenciadas con @nombre.

| Variable                | Descripción                     |
| ----------------------- | ------------------------------- |
| `@background`           | Fondo principal                 |
| `@foreground`           | Texto principal                 |
| `@accentColor`          | Color de acento global          |
| `@selectionBackground`  | Fondo al seleccionar            |
| `@selectionForeground`  | Texto seleccionado              |
| `@disabledForeground`   | Texto deshabilitado             |
| `@disabledBackground`   | Fondo deshabilitado             |
| `@componentBackground`  | Fondo estándar de componentes   |
| `@componentBorderColor` | Color de borde de componentes   |
| `@infoBackground`       | Fondo de información            |
| `@warningBackground`    | Fondo de advertencia            |
| `@errorBackground`      | Fondo de error                  |
| `@focusColor`           | Color de foco general           |
| `@hoverBackground`      | Fondo general al pasar el mouse |
